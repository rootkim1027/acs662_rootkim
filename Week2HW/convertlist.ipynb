```python
import numpy as np
import math
from math import pi
def convert(x,y):
    r=(x**2+y**2)**0.5
    if x>0 and y>0 :
        th=(math.atan(y/x))
    elif x>0 and y<0 :
        th=(math.atan(y/x)+2*pi)
    elif x<0 :
        th=(math.atan(y/x)+pi)
    elif x==0 and y>0 :
        th=pi/2
    else :
        th=(3*pi)/2
    return([r,th])

def convertlist(n)   :
    pr = [] 
    pth = [] 
    sx=np.random.normal(0,5,(1,n))
    sy=np.random.normal(0,5,(1,n))
    a=sx[x]
    b=sy[y]
    k=0
    while k<n :
        c=convert(a[k],b[k])
        pr.insert(k,c[0])
        pth.insert(k,c[1])
        k=k+1
    print(pr[:])
    print(pth[:])

convertlist(100)


    
```
